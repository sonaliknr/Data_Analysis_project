```python
#Renaming of columns element 
data.columns = (["country","region","population","area","density","coastline","migration","infant_mortality","gdp","literacy","phones","arable","crops","other","climate","birthrate","deathrate","agriculture","industry","service"])
```

```python
#head gives the information of dataset after renaming
data.head()
```

```python
#We have to change our datatype to "category" and "float" to use easily and replace , with .
data.country = data.country.astype('category')
data.region = data.region.astype('category')
data.density = data.density.str.replace(",",".").astype(float)
data.coastline = data.coastline.str.replace(",",".").astype(float)
data.migration = data.migration.str.replace(",",".").astype(float)
data.infant_mortality = data.infant_mortality.str.replace(",",".").astype(float)
data.literacy = data.literacy.str.replace(",",".").astype(float)
data.phones = data.phones.str.replace(",",".").astype(float)
data.arable = data.arable.str.replace(",",".").astype(float)
data.crops = data.crops.str.replace(",",".").astype(float)
data.other = data.other.str.replace(",",".").astype(float)
data.climate = data.climate.str.replace(",",".").astype(float)
data.birthrate = data.birthrate.str.replace(",",".").astype(float)
data.deathrate = data.deathrate.str.replace(",",".").astype(float)
data.agriculture = data.agriculture.str.replace(",",".").astype(float)
data.industry = data.industry.str.replace(",",".").astype(float)
data.service = data.service.str.replace(",",".").astype(float)
```

```python
data.info()
```

Yes our datatypes are perfect now. We can go on. 

```python
#After replacing , with . our data
data.head()
```

There are some Nan values and these will be a problem for us. So we have fix them. Now Let's learn how many Nan values in our dataset.

```python
missing = data.isnull().sum()
missing
```

Let's fill the missing values by using fillna method.We will fill Nan values with mean of columns

```python
data.fillna(data.mean(),inplace=True)
```

```python
data.info()
```

```python
data
```

```python
# it calculating total no of NaN value in each column
data.isnull().sum()
```

In the region column , There are some spaces before and after of the regions.We can fix this issue 

```python
data['region']=data['region'].str.replace(' ','')
```

Data visualization is the presentation of data in a pictorial or graphical format. It enables decision makers to see analytics presented visually, so they can grasp difficult concepts or identify new patterns. NOW Let's make some visualization

```python
#correlation map
#annot = True means we can see numbers in square, fmt = number of digits after comma
f,ax = plt.subplots(figsize = (18,16))
sns.heatmap(data.corr(),annot = True, linewidth = 8, fmt = ".1f", ax=ax)  #seaborn 
plt.show()
```

```python
sns.countplot(y='region',data=data)
plt.title('Number of Countries by REGİON',color = 'red',fontsize=15)
```

As you see on graph , number one region is Sub-saharan Africa.

```python
'''BoxPlot'''

plt.subplots(figsize=(18,6))
sns.boxplot(x='region',y='population',data=data)
plt.xticks(rotation=45)
plt.title('Population by Region',color='blue',fontsize=20)
```
